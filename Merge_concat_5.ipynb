# Merge

```python
import pandas as pd
import numpy as np
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,5], "C":[21,22,23,24,25]})
pd.merge(var1,var2,on="A")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,5], "C":[21,22,23,24,25]})
pd.merge(var2,var1,on="A")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6], "C":[21,22,23,24,25]})
pd.merge(var2,var1,on="A")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6], "C":[21,22,23,24,25]})
pd.merge(var2,var1,how="inner")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6], "C":[21,22,23,24,25]})
pd.merge(var2,var1,how="left")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6], "C":[21,22,23,24,25]})
pd.merge(var2,var1,how="right")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6], "C":[21,22,23,24,25]})
pd.merge(var2,var1,how="outer")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6], "C":[21,22,23,24,25]})
pd.merge(var2,var1,how="left",indicator=True)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6], "B":[21,22,23,24,25]})
pd.merge(var2,var1,left_index=True,right_index=True)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6], "B":[21,22,23,24,25]})
pd.merge(var2,var1,left_index=True,right_index=True,suffixes=("name","id"))
```

# Concat

```python
sr1= pd.Series([1,2,3,4,5])
sr2= pd.Series([11,12,13,14,15])

pd.concat([sr1,sr2])
```

```python
d12 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
d11 = pd.DataFrame({"A":[31,32,33,34,36], "B":[21,22,23,24,25]})
pd.concat([d12,d11])
```

```python
d12 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
d11 = pd.DataFrame({"A":[31,32,33,34,36], "B":[21,22,23,24,25]})
pd.concat([d12,d11],axis=0)
```

```python
d12 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
d11 = pd.DataFrame({"A":[31,32,33,34,36], "B":[21,22,23,24,25]})
pd.concat([d12,d11],axis=1)
```

```python
d12 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
d11 = pd.DataFrame({"s":[31,32,33,34,36], "c":[21,22,23,24,25]})
pd.concat([d12,d11],axis=1)
```

```python
d12 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
d11 = pd.DataFrame({"s":[31,32], "c":[21,22]})
pd.concat([d12,d11],axis=1)
```

```python
d12 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
d11 = pd.DataFrame({"s":[31,32], "c":[21,22]})
pd.concat([d12,d11],axis=1,join="outer")
```

```python
d12 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
d11 = pd.DataFrame({"s":[31,32], "c":[21,22]})
pd.concat([d12,d11],axis=1,join="inner")
```

```python
d12 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
d11 = pd.DataFrame({"s":[31,32,33,34,35], "c":[21,22,23,24,25]})
pd.concat([d12,d11],axis=0,keys=["d21","d11"])
```

```python
d12 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
d11 = pd.DataFrame({"s":[31,32,33,34,35], "c":[21,22,23,24,25]})
pd.concat([d12,d11],axis=1,keys=["d12","d11"])
```

```python
sr1= pd.Series([1,2,3,4,5])
sr2= pd.Series([11,12,13,14,15])

pd.concat([sr1,sr2],axis=0,keys=["d","g"])
```

```python
sr1= pd.Series([1,2,3,4,5])
sr2= pd.Series([11,12,13,14,15])

pd.concat([sr1,sr2],axis=1,keys=["d","g"])
```

```python
sr1= pd.DataFrame({"A":[1,2,3,4,5]})
sr2= pd.DataFrame({"B":[11,12,13,14,15],"V":[21,22,23,24,24]})

pd.concat([sr1,sr2])
```
