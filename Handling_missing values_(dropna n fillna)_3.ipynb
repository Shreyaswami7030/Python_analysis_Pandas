```python
import pandas as pd
import numpy as np

# Example DataFrame
data = {
    'A': [1, np.nan, 3, np.nan],
    'B': [4, 5, np.nan, np.nan],
    'C': [7, 8, 9, np.nan],
    'D': [np.nan, 11, 12, 13]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Using dropna with thresh=3
result = df.dropna(thresh=2)
print("\nDataFrame after applying dropna(thresh=3):")
print(result)
```

```python
import pandas as pd
import numpy as np

# Create a DataFrame
data = {
    'A': [1, np.nan, 3, np.nan, 5],
    'B': [4, 5, np.nan, np.nan, 9],
    'C': [7, 8, 9, np.nan, 10],
    'D': [np.nan, 11, 12, 13, 14],
    'E': [np.nan, np.nan, np.nan, 15, 16]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Apply dropna with thresh=5
result = df.dropna(thresh=5)
print("\nDataFrame after dropna(thresh=5):")
print(result)
```
